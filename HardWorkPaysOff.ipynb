```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  
```

# Section 1. Load & Visualize the Data
- Download
- Load
- Visualise

```python
X = pd.read_csv('/Users/varun/Desktop/ML & AI/ML_CodingBlocks/Linear_X_Train.csv')
y = pd.read_csv('/Users/varun/Desktop/ML & AI/ML_CodingBlocks/Linear_Y_Train.csv')

# Convert X,Y to Numpy arrays
X = X.values
y = y.values
# Visualise
plt.style.use('seaborn')
plt.scatter(X,y,color='orange')
plt.title("Hardwork vs Performance Graph")
plt.xlabel("Hardwork")
plt.ylabel("Performance")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize=True)
lr.fit(X,y)
```

```python
lr.predict([[10]])
```

```python
plt.scatter(X,y,color='orange')
plt.plot(X,lr.predict(X),color='Black')
```

```python
from sklearn.externals import joblib
```
